```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
fig = plt.figure(figsize=(6, 6), dpi=100)

ax = fig.add_subplot(1, 1, 1)

ax.xaxis.set_major_locator(MultipleLocator(90))
ax.yaxis.set_major_locator(MultipleLocator(1))

ax.set_xlim(0,360)
ax.set_ylim(-1,1)

mid_x = (ax.get_xlim()[1] + ax.get_xlim()[0])/2
mid_y = (ax.get_ylim()[1] + ax.get_ylim()[0])/2

plt.hlines(mid_y, ax.get_xlim()[0], ax.get_xlim()[1], color='black', linestyle='dotted', zorder=-2,lw=0.5)
plt.vlines(mid_x, ax.get_ylim()[0], ax.get_ylim()[1], color='black', linestyle='dotted', zorder=-2,lw=0.5)


fig.set_facecolor('lightgray')
ax.set_facecolor('lightblue')

fig_width, fig_height = fig.get_size_inches()
fig_aspect_ratio = fig_height / fig_width


fig.lines.append(
    plt.Line2D(
        [0, 1],        # x-coordina tes in figure space
        [0, fig_aspect_ratio],  # y-coordinates in figure space
        transform=fig.transFigure,
        color='black',
        zorder=2,
        alpha=0.95
    )
)


fig.lines.append(
    plt.Line2D(
        [1, 0],        # x-coordinates in figure space
        [0, fig_aspect_ratio],  # y-coordinates in figure space
        transform=fig.transFigure,
        color='black',
        zorder=2,
        alpha=0.5
    )
)

#ax.set_position([0.2, 0.2, 1, 1])
plt.show()
```
