```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter

np.random.seed(123)

X = np.linspace(0.5, 3.5, 100)
Y1 = 3 + np.cos(X)
Y2 = 1 + np.cos(1 + X / 0.75) / 2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)


def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x


ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which="major", width=1.0)
ax.tick_params(which="major", length=10)
ax.tick_params(which="minor", width=1.0, labelsize=10)
ax.tick_params(which="minor", length=5, labelsize=10, labelcolor="0.25")

ax.grid(linestyle="--", linewidth=0.5, color=".25", zorder=-10)

ax.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(X, Y3, linewidth=0, marker="o", markerfacecolor="w", markeredgecolor="k")

ax.set_title("Anatomy of a figure", fontsize=20, verticalalignment="bottom")
ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")

ax.legend()


def circle(x, y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke

    circle = Circle(
        (x, y),
        radius,
        clip_on=False,
        zorder=10,
        linewidth=1,
        edgecolor="black",
        facecolor=(0, 0, 0, 0.0125),
        path_effects=[withStroke(linewidth=5, foreground="w")],
    )
    ax.add_artist(circle)


def text(x, y, text):
    ax.text(
        x,
        y,
        text,
        backgroundcolor="white",
        # fontname="Yanone Kaffeesatz", fontsize="large",
        ha="center",
        va="top",
        weight="regular",
        color="#000099",
    )


# Minor tick
circle(0.50, -0.10)
text(0.50, -0.32, "Minor tick label")

# Major tick
circle(-0.03, 4.00)
text(0.03, 3.80, "Major tick")

# Minor tick
circle(0.00, 3.50)
text(0.00, 3.30, "Minor tick")

# Major tick label
circle(-0.15, 3.00)
text(-0.15, 2.80, "Major tick label")

# X Label
circle(1.80, -0.27)
text(1.80, -0.45, "X axis label")

# Y Label
circle(-0.27, 1.80)
text(-0.27, 1.6, "Y axis label")

# Title
circle(1.60, 4.13)
text(1.60, 3.93, "Title")

# Blue plot
circle(1.75, 2.80)
text(1.75, 2.60, "Line\n(line plot)")

# Red plot
circle(1.20, 0.60)
text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(3.20, 1.75)
text(3.20, 1.55, "Markers\n(scatter plot)")

# Grid
circle(3.00, 3.00)
text(3.00, 2.80, "Grid")

# Legend
circle(3.70, 3.80)
text(3.70, 3.60, "Legend")

# Axes
circle(0.5, 0.5)
text(0.5, 0.3, "Axes")

# Figure
circle(-0.3, 0.65)
text(-0.3, 0.45, "Figure")

color = "#000099"
ax.annotate(
    "Spines",
    xy=(4.0, 0.35),
    xytext=(3.3, 0.5),
    color=color,
    weight="regular",  # fontsize="large", fontname="Yanone Kaffeesatz",
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color=color),
)

ax.annotate(
    "",
    xy=(3.15, 0.0),
    xytext=(3.45, 0.45),
    color=color,
    weight="regular",  # fontsize="large", fontname="Yanone Kaffeesatz",
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color=color),
)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter
```

```python
## Generate random dataset (X) and (Y1, Y2, Y3) which are derive

np.random.seed(123)

X = np.linspace(0.5, 3.5, 100)
Y1 = 3 + np.cos(X)
Y2 = 1 + np.cos(1 + X / 0.75) / 2
Y3 = np.random.uniform(Y1, Y2, len(X))

```

```python
## Instantiate a figure and an axis object

"""
The plt.figure() method creates and returns a Figure object, which is the overall container for all plot elements (axes, titles, labels, etc.).
'figsize' specifies the width and height of the figure in inches.
"""
fig = plt.figure(figsize=(8, 8))


"""
We can add a subplot (or an axis) to the figure using the add_subplot() method.
The add_subplot() method returns an Axes object, which acts as the canvas for plotting elements (lines, markers, etc.).
The arguments (1, 1, 1) specify a grid of 1 row and 1 column, and the subplot is placed in position 1 (only one subplot here).
'aspect=1' sets the aspect ratio to be equal, so the units are the same on both axes.
"""

ax = fig.add_subplot(1, 1, 1, aspect=1)
```

```python
"""
Custom formatting function for minor ticks.
Checks if the tick is an integer and returns an empty string if it is.
Otherwise, it returns the tick value formatted to two decimal places.

For example, 1 will be formatted as ''.
For example, 0.5 will be formatted as '0.50'.

the pos argument is not used but required by matplotlib when defining custom tick formatting.

"""
def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x


minor_tick(1, None),\
    minor_tick(0.5, None)
```

```python
ax.xaxis.set_major_locator(MultipleLocator(0.1))
ax.yaxis.set_major_locator(MultipleLocator(0.1))
ax.figure
```

```python
"""
The MultipleLocator class is used to set tick intervals on an axis.
ets the tick marks at intervals of 1.0.
The locator will place ticks at multiples of the specified value (in this case, every 1.0 unit).
So the x-axis will have ticks at positions like 0, 1, 2, 3, 4, ... and so on.

Used with set_major_locator() or set_minor_locator() to control the placement of major or minor ticks.

"""
ax.xaxis.set_major_locator(MultipleLocator(0.1))
ax.yaxis.set_major_locator(MultipleLocator(0.1))
ax.figure
```

```python
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(1))
ax.figure
```

```python
"""

AutoMinorLocator automatically determines the positions of minor ticks based on the major tick intervals.

Automatically places minor ticks between the major ticks.
The number of minor ticks is chosen based on the major tick intervals.
By default, it places 4 minor ticks between each pair of major ticks.

This only places the ticks, which are the small lines on the axis. It does not label the ticks.
"""
```

```python
ax.xaxis.set_minor_locator(AutoMinorLocator(8))
ax.yaxis.set_minor_locator(AutoMinorLocator(8))
ax.figure
```

```python
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.figure
```

```python
"""
minor_tick() was defined above as a custom formatting function for minor ticks.

There are a different built-in formatters available in matplotlib.ticker module, for example, PercentFormatter, ScalarFormatter, etc.

"""

from matplotlib.ticker import PercentFormatter
ax.xaxis.set_minor_formatter(PercentFormatter(xmax=1.0))
ax.figure
```

```python
from matplotlib.ticker import StrMethodFormatter
ax.xaxis.set_minor_formatter(StrMethodFormatter("{x:.2f}"))
ax.figure
```

```python
"""
minor_tick() was defined above as a custom formatting function for minor ticks.

FuncFormatter is a class that formats tick labels using a user-defined function.

set_minor_formatter() is used to set a custom formatting function for minor ticks.

"""

ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))
ax.figure
```

```python

"""
set_xlim() and set_ylim() methods are used to set the limits of the x-axis and y-axis, respectively.
By default, when setting the axis with fig.add_subplot(1, 1, 1, aspect=1), the limits are set to 0 to 1.

The minor tick labels are placed on the x-axis only since we only applied set_minor_formatter on the xaxis attribute

"""

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
ax.figure
```

```python
"""
tick_params() controls the appearance of ticks on the axes.

The 'which' argument specifies whether to apply the settings to major ticks or minor ticks.
The 'width' argument sets the width of the ticks.
The 'length' argument sets the length of the ticks.
The 'labelsize' argument sets the font size of the tick labels.
The 'labelcolor' argument sets the color of the tick labels.
labelcolor accepts a string color name, a string color code, or a tuple of RGB values.

"""

ax.tick_params(which="major", width=1.0)
ax.tick_params(which="major", length=10)
ax.tick_params(which="minor", width=2.0, labelsize=10)
ax.tick_params(which="minor", length=5, labelsize=8, labelcolor="red")
ax.figure
```

```python
ax.xaxis.get_tick_params(which="major"),\
    ax.xaxis.get_tick_params(which="minor")
```

```python
"""

The .grid() method is used to display gridlines on the plot.
Here we set a linestyle of '--' (dashed line), a linewidth of 0.5, colored red.

zorder is used to control the order of the gridlines with respect to other plot elements.
zorder=-10 places the gridlines behind the plot elements.

"""
ax.grid(linestyle="--", linewidth=0.5, color='red', zorder=-10)
ax.figure
```

```python
"""
Finally we use the plot() method to plot the three lines.

The 'c' argument specifies the color of the line.
The 'lw' argument specifies the line width.
The 'label' argument specifies the label for the line in the legend.
The 'zorder' argument specifies the order of the line with respect to other plot elements.
zorder=10 places the line ON TOP of the gridlines.

"""


ax.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
ax.figure
```

```python
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.figure
```

```python
""" 
Here we plot the third dataset which are random points, not a line.
Matplotlib will automatically connect the points with straight lines unless specified otherwise.
If a marker is specified, the points will be plotted as markers without connecting lines.
"""


ax.plot(X, Y3, linewidth=0, marker="o", markerfacecolor="w", markeredgecolor="k")
ax.figure
```

```python
""" 
.set_title() method is used to set the title of the plot.
    The 'fontsize' argument specifies the font size of the title.
    The 'verticalalignment' argument specifies the vertical alignment of the title.
.set_xlabel() and .set_ylabel() methods are used to set the labels for the x-axis and y-axis, respectively.

"""

ax.set_title("Anatomy of a figure", fontsize=20, verticalalignment="bottom")
ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")
ax.figure
```

```python
"""
The .legend() method is used to display the legend, which shows the labels of the lines on the plot.
By default, the legend is placed in the best location to avoid overlapping with the plot elements.
"""

ax.legend()
ax.figure

```

```python

"""
The circle() function is used to draw a circle on the plot.
It uses the Circle class from matplotlib.patches to create a circle.
The withStroke path effect is used to create a shadow effect around the circle.

The Circle object is added to the axis using the add_artist() method.
An artist is a base class for all the objects that can be added to the plot.

The circle is defined by its center (x, y), radius. Everythihng else is optional.
The clip_on argument controls whether the circle is clipped to the axis limits or not.
Clipping means that the circle will not be drawn outside the axis limits.
If clip_on is False, the circle will be drawn even if it extends beyond the axis limits,
but if it's True, it will be clipped at the axis limits.

The path effects control the appearance of the circle.

withStroke:
Adds a stroke (outline) around a text or shape path, making it easier to read when placed over complex backgrounds or overlapping elements.
The stroke effect typically involves drawing an outline in a different color around the main element (e.g., text or shape).
The foreground argument specifies the color of the outline (stroke).

"""



def circle(x, y, radius=0.15, clip_on=False):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke

    circle = Circle(
        (x, y),
        radius,
        clip_on=clip_on,
        zorder=10,
        linewidth=1,
        edgecolor="black",
        facecolor=(0, 0, 0, 0.0125),
        path_effects=[withStroke(linewidth=5, foreground="w")],
    )
    ax.add_artist(circle)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe

fig, ax3 = plt.subplots()

# Apply a stroke effect using path effects
text1 = ax3.text(0.5, 0.75, "withStroke=False", fontsize=30, ha='center', va='center')

text2 = ax3.text(0.5, 0.25, "withStroke=True", fontsize=30, ha='center', va='center')
text2.set_path_effects([pe.withStroke(linewidth=3, foreground="black")])



ax3.spines[['bottom','left','right', 'top']].set_visible(False)
ax3.set_xticks([])
ax3.set_yticks([])

plt.show()
```
