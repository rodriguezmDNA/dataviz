```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.transforms as transforms
from matplotlib.patches import Circle

fig, ax = plt.subplots(
    subplot_kw={'projection': 'polar'},
    figsize=(5, 5)
)

# Remove extra space around the plot
#plt.subplots_adjust(left=0.1, right=1, top=1, bottom=0)

ax.set_ylim(0, 5)
ax.set_aspect('equal', 'box')  # Helps center the polar plot as a true circle
data_to_display = ax.transData  # transform (theta, r) -> display coords

def draw_same_size_circle(ax, theta, r, radius_pts=6, color='black'):
    # Get the location in "screen" (display) coords
    center_display = data_to_display.transform((theta, r))
    # Make a circle at origin in display coords, with a fixed radius in points
    circle = Circle((0, 0), radius_pts, facecolor=color, clip_on=False)
    # Shift circle's center to the desired location
    circle.set_transform(transforms.IdentityTransform() +
                         transforms.Affine2D().translate(*center_display))
    ax.add_patch(circle)


# Draw circles at angle pi
for y in np.arange(1, 5.5, 0.5):
    draw_same_size_circle(ax, np.pi, y, radius_pts=5, color='black')

# Draw circles at angle 0
for y in np.arange(1, 5.5, 0.5):
    draw_same_size_circle(ax, 0, y, radius_pts=5, color='tab:blue')

# Draw circles at angle pi/4
for y in np.arange(1, 5.5, 0.5):
    draw_same_size_circle(ax, np.pi/4, y, radius_pts=5, color='tab:green')

for y in np.arange(1, 5.5, 0.5):
    draw_same_size_circle(ax, np.pi/2, y, radius_pts=5, color='tab:red')


for y in np.arange(1, 5.5, 0.5):
    draw_same_size_circle(ax, -(3*np.pi)/4, y, radius_pts=5, color='tab:green')


ax.set_facecolor((0,0,0,0))

fig.tight_layout()

fig_width, fig_height = fig.get_size_inches()
fig_aspect_ratio = fig_height / fig_width


fig.lines.append(
    plt.Line2D(
        [0, 1],        # x-coordinates in figure space
        [0, fig_aspect_ratio],  # y-coordinates in figure space
        transform=fig.transFigure,
        color='black',
        zorder=-2,
        alpha=0.95
    )
)


fig.lines.append(
    plt.Line2D(
        [1, 0],        # x-coordinates in figure space
        [0, fig_aspect_ratio],  # y-coordinates in figure space
        transform=fig.transFigure,
        color='black',
        zorder=-2,
        alpha=0.95
    )
)

plt.show()
```
