```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
import textwrap
```

```python
#Marker size is expressed in point^2 and a point is defined as fig.dpi/72


X = 0.5 + np.arange(8)
Y = np.ones(len(X))

fig1 = plt.figure(dpi=100)  # Adjust DPI as needed
ax1 = fig1.add_subplot(111)
scatter1 = ax1.scatter(X, Y, s=100, facecolor="none", edgecolor="black", linewidth=1)
plt.show()

fig2 = plt.figure(dpi=200)  # Adjust DPI as needed
ax2 = fig2.add_subplot(111)
scatter2 = ax2.scatter(X, Y, s=100, facecolor="none", edgecolor="black", linewidth=1)
plt.show()
```

```python
import math 
point_size = scatter1.get_sizes()[0]  # 100

# Calculate the physical size in inches
physical_size = math.sqrt(point_size) / 72
print(f"Physical size of markers: {physical_size:.4f} inches")

point_size = scatter2.get_sizes()[0]  # 100

# Calculate the physical size in inches
physical_size = math.sqrt(point_size) / 72
print(f"Physical size of markers: {physical_size:.4f} inches")
```

### Pixels per inch

The number of inches corresponding to a given number of pixels is $\frac{pixels}{dpi}$

For any DPI the inches remain the same but the density of pixels increases.
* DPI = 100, 100 pixels per inch
* DPI = 200, 200 pixels per inch

### Points

Regardless of the dpi, 1 inch = 72 points. 

For a given dpi the number of pixels per point is determined by $\frac{pixels}{72 points}$

* For dpi = 100, 100/72 = 1.38 pixels per point
* For dpi = 200, 200/72 = 2.77 pixels per point

```python
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


#######################################
fig = plt.figure(figsize=(6, 6), dpi=100)


ax1 = fig.add_subplot(1, 1, 1)
ax1.set_xlim(0,360)
ax1.set_ylim(0,1)


# Get the position of ax1 in figure coordinates
bbox = ax1.get_position()

# Compute ax2 position in figure coordinates based on ax1's axes coordinates

# Define ax2 with the calculated figure coordinates
ax2 = fig.add_axes( [bbox.x0 + 0.65 * bbox.width,
                     bbox.y0 + 0.65 * bbox.height,
                     .35 * bbox.width,
                     .35 * bbox.height
                     ],
                   aspect=1,
                   xlim=[0, 10], ylim=[0, 10])

ax1.xaxis.set_major_locator(MultipleLocator(90))
ax1.yaxis.set_major_locator(MultipleLocator(0.1))


ax2.xaxis.set_major_locator(MultipleLocator(9))
ax2.yaxis.set_major_locator(MultipleLocator(1))


# Draw bounding box of ax1
ax1.add_patch(Rectangle((0, 0), 1, 1, transform=ax1.transAxes, clip_on=False,
                        fill=False, edgecolor="red", linestyle="--", lw=2))

# Draw bounding box of ax2
ax2.add_patch(Rectangle((0, 0), 1, 1, transform=ax2.transAxes, clip_on=False,
                        fill=False, edgecolor="blue", linestyle="--", lw=2))


ax1.hlines(0.5, 0, 1, color='tab:blue', linestyle='dotted', zorder=1,lw=1,transform=ax1.transAxes) 
ax1.hlines(0.7, 0, 1, color='tab:orange', linestyle='-', zorder=1,lw=1,transform=fig.transFigure) 


ax2.hlines(0.5, 0, 1, color='tab:blue', linestyle='dotted', zorder=1,lw=1,transform=ax2.transAxes) 
ax2.hlines(0.75, 0, 1, color='tab:red', linestyle='dotted', zorder=1,lw=1,transform=ax2.transAxes) 
ax2.hlines(0.7, 0, 1, color='black', linestyle='dotted', zorder=1,lw=1,transform=fig.transFigure) 


fig.set_facecolor('lightgray')
ax1.set_facecolor('lightblue')
ax2.set_zorder(1)
ax2.set_facecolor(mcolors.to_rgba('white', alpha=0.15))

plt.show()
```
