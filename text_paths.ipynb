```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.textpath import TextPath
from matplotlib.path import Path
from matplotlib.patches import PathPatch
from matplotlib.font_manager import FontProperties

def interpolate(X, Y, T):
    if len(X) < 2 or len(Y) < 2:
        return None, None, 0
    dR = np.sqrt(np.diff(X)**2 + np.diff(Y)**2)
    R = np.zeros_like(X)
    R[1:] = np.cumsum(dR)
    if R[-1] == 0:
        return None, None, 0
    return (
        np.interp(T, R, X),
        np.interp(T, R, Y),
        R[-1]
    )

def contour_text(X, Y, text, offset=0, text_size=1.0, scale=1.0):
    """
    Draws text bent along a contour.
      text_size: set the font size (smaller = smaller glyphs)
      scale    : an extra scale factor on top of text_size 
    """
    if len(X) < 2 or len(Y) < 2:
        return
    
    # Create a TextPath in local (0,0) coords, smaller if desired
    text_path = TextPath(
        (0, -0.75), text,
        prop=FontProperties(size=text_size, family="Roboto", weight="bold")
    )
    # Copy glyph geometry & codes
    glyph_vertices = text_path.vertices.copy()
    glyph_codes    = text_path.codes
    
    # Optionally scale down glyphs more
    glyph_vertices *= scale
    
    # Interpolate anchor positions for glyph "x" coords
    out = interpolate(X, Y, offset + glyph_vertices[:, 0])
    if out[0] is None:
        return
    X0, Y0, total_length = out

    # A second interpolation for direction
    out = interpolate(X, Y, offset + glyph_vertices[:, 0] + 0.1)
    if out[0] is None:
        return
    X1, Y1, _ = out

    # Plot leftover path
    out = interpolate(X, Y, np.linspace(glyph_vertices[:, 0].max() + 1,
                                        total_length - 1, 200))
    if out[0] is not None:
        Xr, Yr, _ = out
        plt.plot(Xr, Yr, color="black", linewidth=0.5,
                 markersize=1, marker="o", markevery=[0, -1])
    
    # Adjust glyph vertices to follow the curve
    dX, dY = (X1 - X0) + 1e-6, (Y1 - Y0) + 1e-6
    norm = np.sqrt(dX**2 + dY**2)
    dX, dY = dX / norm, dY / norm

    # 'Bend' the text by shifting in local y-direction
    X0 += -glyph_vertices[:, 1] * dY
    Y0 +=  glyph_vertices[:, 1] * dX
    
    new_vertices = np.column_stack((X0, Y0))
    new_path = Path(new_vertices, glyph_codes)

    ax = plt.gca()
    patch_bg = PathPatch(new_path, facecolor="white", edgecolor="white",
                         linewidth=1.25, alpha=0.25, zorder=10)
    patch_bg.set_transform(ax.transData)
    ax.add_artist(patch_bg)

    patch_fg = PathPatch(new_path, facecolor="black", edgecolor="black",
                         linewidth=0.0, zorder=30)
    patch_fg.set_transform(ax.transData)
    ax.add_artist(patch_fg)

# Example usage
n = 100
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)
Z = 0.5 * (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig = plt.figure(figsize=(10,5), dpi=100)
levels = 10

# Left: normal labels
ax1 = fig.add_subplot(1, 2, 1, aspect=1, xticks=[], yticks=[])
cf1 = ax1.contourf(x, y, Z, levels=levels)
cs1 = ax1.contour(x, y, Z, levels=levels, colors="black", linewidths=0.5)
ax1.clabel(cs1, cs1.levels)
ax1.set_title("Default Contour Labels")

# Right: curved labels, smaller text
ax2 = fig.add_subplot(1, 2, 2, aspect=1, xticks=[], yticks=[])
cf2 = ax2.contourf(x, y, Z, levels=levels)
cs2 = ax2.contour(x, y, Z, levels=levels, alpha=0, colors="black", linewidths=0.5)

for level, segments in zip(cs2.levels, cs2.allsegs):
    for seg in segments:
        if len(seg) < 2:
            continue
        seg = np.array(seg)
        label = f"{level:.3f}"
        if level == 0.0:
            label = " DO NOT CROSS " * 8
        # Use a smaller font and/or scale factor
        contour_text(seg[:,0], seg[:,1], label, text_size=0.85, scale=0.1)

ax2.set_title("Curved Contour Labels (Smaller Text)")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.textpath import TextPath
from matplotlib.patches import PathPatch
from matplotlib.path import Path
import numpy as np
```

```python
fig = plt.figure(figsize=(4.25, 2))
ax = fig.add_axes([0, 0, 1, 1], aspect=1, xlim=[-40, 40], ylim=[-1, 25])
ax.axis("off")

text = [
    "Beautiful is better than ugly.",
    "Explicit is better than implicit.",
    "Simple is better than complex.",
    "Complex is better than complicated.",
    "Flat is better than nested.",
    "Sparse is better than dense.",
    "Readability counts.",
]

y = 0
size = 6
xfactor = 1 / 50
yfactor = 1 / 120

for i, line in enumerate(text[::-1]):
    path = TextPath((0, 0), line, size=size)
    V = path.vertices
    codes = path.codes
    xmin, xmax = V[:, 0].min(), V[:, 0].max()
    ymin, ymax = V[:, 1].min(), V[:, 1].max()

    # New Path
    # X centering
    Px = V[:, 0] - (xmax + xmin) / 2

    # Moving whole text at y coordinates
    Py = V[:, 1] + y

    # Rescaling along y
    Py *= 1 - (Py * yfactor)

    # Rescaling along x
    Px *= 1 - (Py * xfactor)

    # Update interlines
    y += size * (1 - ymin * yfactor)

    # new path
    path_new = Path(np.array([Px, Py]).T, codes=codes)
    
    # Display
    patch = PathPatch(path_new, facecolor="%.2f" % (i / 10), linewidth=0, clip_on=False)
    ax.add_artist(patch)
```
