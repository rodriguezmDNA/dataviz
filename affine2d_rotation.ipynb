```python
# Creating histograms
fig, ax = plt.subplots(figsize=(5, 5))  # Increase figure size for better visibility


ax.hist(normal_data, bins=10,transform=ax.transData, color='tab:gray',clip_on=False)

fig.lines.extend([plt.Line2D([0, 1], [0, 0],
                             transform=fig.transFigure,
                             lw=2,
                             color='tab:gray')
                             ])


transform = Affine2D().rotate_deg(45) + ax.transData
ax.hist(normal_data, bins=10,transform=transform, color='tab:green',clip_on=False)
fig.lines.extend([plt.Line2D([0, 1], [0, 0],
                             transform=Affine2D().rotate_deg(45) + fig.transFigure,
                             lw=2,
                             color='tab:green')
                             ])


transform = Affine2D().rotate_deg(-45) + ax.transData
ax.hist(normal_data, bins=10,transform=transform, color='tab:orange',clip_on=False)
fig.lines.extend([plt.Line2D([0, 1], [0, 0],
                             transform=Affine2D().rotate_deg(-45) + fig.transFigure,
                             lw=2,
                             color='tab:orange')
                             ])

ax.plot([0, 1], [0, 1], transform=ax.transAxes,color='tab:gray')
ax.plot([1, 0], [0, 1], transform=ax.transAxes,color='tab:gray')




fig.set_facecolor('none')
fig.lines.extend([
    plt.Line2D([0, 1], [0, 1], transform=fig.transFigure, color='black', lw=0.5, linestyle='dotted', alpha=1),
    plt.Line2D([0, 1], [1, 0], transform=fig.transFigure, color='black', lw=0.5, linestyle='dotted', alpha=1),
])

ax.set_aspect('equal')

# Save image to pdf
plt.show()
```

```python
# Creating histograms
fig, ax = plt.subplots(figsize=(5, 5))  # Increase figure size for better visibility


increase = 45
for degree in range(0, 360+increase, increase):

    transform = Affine2D().rotate_deg(degree) + ax.transData
    _,_, _patches = ax.hist(normal_data, bins=10,transform=transform, clip_on=False)
    fig.lines.extend([plt.Line2D([0, 1], [0, 0],
                                transform=Affine2D().rotate_deg(degree) + fig.transFigure,
                                lw=2,
                                color=_patches[0].get_facecolor()
                                )
                                ])

    ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))  # Increase figure size for better visibility

increase = 45
for degree in range(0, 360+increase, increase):

    transform = Affine2D().rotate_deg_around(normal_data.mean(),
                                             0,
                                             degree) + ax.transData
    ax.hist(normal_data, bins=10,transform=transform, clip_on=False,alpha=0.65)

ax.set_aspect('equal')
```

```python
# Creating histograms
#fig, ax1, ax2 = plt.subplots(2, figsize=(5, 5))  # Increase figure size for better visibility

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(7, 5))  # Increase figure size for better visibility

ax1_n, ax1_bins, ax1_patches = ax1.hist(normal_data,
                                        bins=10,
                                        transform=ax1.transData,
                                        facecolor='tab:gray',
                                        edgecolor='black',
                                        clip_on=False)

for i, _patch in enumerate(ax1_patches):
    x,y = _patch.get_center()
    y = ax1_n[i]
    y = int(y)
    plt.text(x, y + 5, f'{y}',
             fontsize=8,
             color='black',
             ha='center',
             va='top',
             rotation=0,
             clip_on=False,
             transform=ax1.transData)


transform = Affine2D().rotate_deg_around(np.mean(normal_data), 0, -15) + ax2.transData
ax2_n, ax2_bins, ax2_patches = ax2.hist(normal_data, bins=10,transform=transform, color='tab:green',clip_on=False,alpha=0.65)

for i, _patch in enumerate(ax2_patches):
    x,y = _patch.get_center()
    y = ax2_n[i]
    y = int(y)
    plt.text(x, y + 6, f'{y}',
             fontsize=8,
             color='black',
             ha='center',
             va='top',
             rotation=-15,
             clip_on=False,
             transform=transform)
    

transform = Affine2D().rotate_deg_around(np.mean(normal_data), 0, 180) + ax1.transData
ax3_n, ax3_bins, ax3_patches = ax3.hist(normal_data, bins=10,transform=transform, color='tab:green',clip_on=False,alpha=0.65)
ax3.invert_xaxis()

for i, _patch in enumerate(ax3_patches):
    x,y = _patch.get_center()
    y = ax3_n[i]
    y = int(y)
    plt.text(x, y + 6, f'{y}',
             fontsize=8,
             color='black',
             ha='center',
             va='top',
             rotation=180,
             clip_on=False,
             transform=transform)



for spine in ax1.spines.values():
    spine.set_visible(False)

for spine in ax2.spines.values():
    spine.set_visible(False)

ax1.set_aspect('equal')
ax1.set_xticks([])
ax1.set_yticks([])

ax2.set_aspect('equal')
ax2.set_xticks([])
ax2.set_yticks([])

ax3.set_aspect('equal')
ax3.set_xticks([])
ax3.set_yticks([])


fig.tight_layout()
```
