```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
import matplotlib.colors as mcolors
```

```python
fig = plt.figure(figsize=(6, 6), dpi=100)

ax1 = fig.add_subplot(1, 1, 1)

ax1.set_xlim(0, 360)
ax1.set_ylim(0, 1)

ax1.xaxis.set_major_locator(MultipleLocator(90))
ax1.yaxis.set_major_locator(MultipleLocator(0.25))


# Add inset within ax1
ax2 = ax1.inset_axes([0.65, 0.65, 0.35, 0.35],
                     xlim=(0, 10), ylim=(0, 10),
                     transform=ax1.transAxes)


ax3 = ax1.inset_axes([0.0, 0.0, 90, 0.25],
                     xlim=(0, 90), ylim=(0, 0.25),
                     transform=ax1.transData)


ax1.hlines(0.65, 0, 1, color='tab:gray', linestyle='dotted', zorder=0,lw=0.5,transform=fig.transFigure) 
#ax1.text( 0.15, 0.65, f"65% line ",transform=fig.transFigure)             
# Annotate with an arrow pointing to the line
ax1.annotate("65% line in fig coords",
    xy=(0.2, 0.65),  # Target point
    xytext=(0.27, 0.75),
    xycoords=fig.transFigure,  # Coordinate system for the target point
    textcoords=fig.transFigure,  # Coordinate system for the target point
    ha='center', 
    va='bottom',  # Change to bottom alignment
    backgroundcolor='white',  # Change to white or remove
    arrowprops=dict(
        facecolor='black',
        shrink=0,  # Remove shrink
        width=0.5,
        headwidth=5
    ))


ax1.annotate("65% line in ax1 coords",
    xy=(0.5, 0.65),  # Target point
    xytext=(0.55, 0.75),
    xycoords=ax1.transAxes,  # Coordinate system for the target point
    textcoords=ax1.transAxes,  # Coordinate system for the target point
    ha='right', 
    va='bottom',  # Change to bottom alignment
    backgroundcolor='white',  # Change to white or remove
    arrowprops=dict(
        facecolor='black',
        shrink=0,  # Remove shrink
        width=0.5,
        headwidth=5
    ))

ax1.hlines(0.65, 0, 1, color='tab:green', linestyle='dotted', zorder=3,lw=5,transform=ax1.transAxes) 
ax1.vlines(0.65, 0, 1, color='tab:green', linestyle='dotted', zorder=3,lw=5,transform=ax1.transAxes) 

fig.set_facecolor('lightgray')
ax1.set_facecolor(mcolors.to_rgba('lightblue', alpha=0.5))
ax2.set_facecolor(mcolors.to_rgba('orange', alpha=0.75))




ax3.xaxis.set_ticks_position('top')
ax3.yaxis.set_ticks_position('right')
ax3.xaxis.set_major_locator(MultipleLocator(90))
ax3.yaxis.set_major_locator(MultipleLocator(0.25))


ax2.xaxis.set_ticks_position('top')
ax2.yaxis.set_ticks_position('right')
ax2.xaxis.set_major_locator(MultipleLocator(5))
ax2.yaxis.set_major_locator(MultipleLocator(5))

plt.show()
```
