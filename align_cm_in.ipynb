```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
inch = 2.54
fig_width = 10.8 / inch  # ~4.252 inches
fig_height = 1.25  #  1.250 inches
margin = 0.125  #  0.125 inches


xmin, xmax = 0, 4
ymin, ymax = 0, 1
```

```python
# Specify the figure size and the margins

fig = plt.figure(figsize=(fig_width, fig_height))

plt.subplots_adjust(
    left=margin / fig_width,
    right=1 - margin / fig_width,
    bottom=margin / fig_height,
    top=1 - margin / fig_height,
)

# Inches graduation
ax1 = plt.subplot(1, 1, 1, yticks=[])
ax1.set_xlim(xmin, xmax)
ax1.set_ylim(ymin, ymax)
ax1.spines['bottom'].set_color('blue')

plt.show()
```

```python
# Inches graduation
ax1 = plt.subplot(1, 1, 1, yticks=[])
ax1.set_xlim(xmin, xmax)
ax1.set_ylim(ymin, ymax)
ax1.spines['bottom'].set_color('blue')
```

```python
# Make the spines invisible except for the bottom spine
# The bottom spine is moved to the middle of the figure
# set_position(("axes", 0.45)) indicates that the spine is at 45% of the figure height. The "xxes" argument indicates that the position is relative to the axes.

ax1.spines["right"].set_visible(False)
ax1.spines["left"].set_visible(False)
ax1.spines["top"].set_visible(False)
ax1.spines["bottom"].set_position(("axes", 0.45))
ax1.spines['bottom'].set_color('blue')
ax1.figure
```

```python
ax1.xaxis.set_major_locator(MultipleLocator(1.00))
ax1.xaxis.set_minor_locator(MultipleLocator(0.25))
ax1.tick_params(axis="both", which="major")
ax1.set_xlabel("inch")
ax1.spines['bottom'].set_color('blue')
ax1.figure
```

```python
# Centimeter graduation
ax2 = ax1.twiny()
ax2.spines['top'].set_color('red')
ax2.figure
```

```python
ax2.spines["right"].set_visible(False)
ax2.spines["left"].set_visible(False)
ax2.spines["bottom"].set_visible(False)
ax2.set_xlim(xmin * inch, xmax * inch)

ax2.figure
```

```python
ax2.spines["top"].set_position(("axes", 0.55))
ax2.figure
```

```python
ax2.xaxis.set_major_locator(MultipleLocator(1.00))
ax2.xaxis.set_minor_locator(MultipleLocator(0.10))
ax2.figure
```

```python
ax2.tick_params(axis="both", which="major", labelsize=10)
ax2.set_xlabel("cm")
ax2.figure
```
