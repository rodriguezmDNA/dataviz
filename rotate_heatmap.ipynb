```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.transforms import Affine2D
import mpl_toolkits.axisartist.floating_axes as floating_axes
```

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Convert to a Pandas DataFrame
iris_df = pd.DataFrame(
    data=iris.data,
    columns=iris.feature_names
)
iris_df
```

```python
corr_ = iris_df.sample(25,random_state=42).T.corr()
corr_ = corr_ ** 2
upper_tri = corr_.where(np.triu(np.ones(corr_.shape), k=1).astype(bool))
#upper_tri = corr_.where(np.tril(np.ones(corr_.shape), k=-1).astype(bool))
upper_tri
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.transforms import Affine2D
import mpl_toolkits.axisartist.floating_axes as floating_axes

# Create figure and sample data
h, w = 12, 12
upper_tri = np.triu(np.random.rand(10, 10))
fig = plt.figure(figsize=(h, w))

# Create base axes
ax1 = fig.add_axes([0.05, 0.05, 0.9, 0.9], aspect=1)
ax1.set_xticklabels([])
ax1.set_yticklabels([])
ax1.xaxis.set_ticks([])
ax1.yaxis.set_ticks([])
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)

# Set up rotated axes
rotation = 45
xmin, xmax = -18, 18
ymin, ymax = 0, xmax - xmin
transform = Affine2D().rotate_deg(rotation-90)
helper = floating_axes.GridHelperCurveLinear(transform, (xmin, xmax, ymin, ymax))

# Create and configure floating subplot
ax2 = floating_axes.FloatingSubplot(fig, 111, grid_helper=helper, zorder=100)
ax2.patch.set_facecolor('green')
ax2.patch.set_alpha(0.3)
fig.add_subplot(ax2)

# Remove all axes elements
for axis in ["left", "right", "top", "bottom"]:
    ax2.axis[axis].set_visible(False)
ax2.set_xticks([])
ax2.set_yticks([])

# Plot heatmap on auxiliary axes
ax2_aux = ax2.get_aux_axes(transform)
sns.heatmap(upper_tri, cmap='Blues', cbar=False, ax=ax2_aux)

plt.show()
```

```python
h, w = 12, 12
fig = plt.figure(figsize=(h, w))

# Define axis range to center the heatmap
xmin, xmax = -10, 10
ymin, ymax = -10, 10

# Adjust transform for rotation
rotation = 45
transform = Affine2D().rotate_deg(rotation)

# Set up grid helper with appropriate limits
helper = floating_axes.GridHelperCurveLinear(transform, (xmin, xmax, ymin, ymax))

# Create a rotated subplot
ax_rotated = floating_axes.FloatingSubplot(fig, 111, grid_helper=helper, zorder=100)
fig.add_subplot(ax_rotated)

# Set a background color for visualization
ax_rotated.patch.set_facecolor('green')
ax_rotated.patch.set_alpha(0.3)

# Remove ticks and spines for a clean look
for axis in ["left", "bottom", "top", "right"]:
    ax_rotated.axis[axis].set_visible(False)

# Get the auxiliary axis and plot the heatmap
ax_rotated_aux = ax_rotated.get_aux_axes(transform)
sns.heatmap(upper_tri, cmap='Blues', cbar=False, ax=ax_rotated_aux)

# Ensure no clipping and proper alignment

#plt.subplots_adjust(left=-1, right=0.9, top=0.9, bottom=0.1)  # Adjust overall figure positioning
plt.show()
```
